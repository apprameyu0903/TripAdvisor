```python
import pandas as pd
```

```python
df=pd.read_csv("zomato.csv")
df.drop(['menu_item','listed_in(type)','listed_in(city)'],axis=1)
df.head()
```

```python
df['dish_liked']=df['dish_liked'].str.split(",",n=0,expand=False)
df['cuisines']=df['cuisines'].str.split(",",n=0,expand=False)
```

```python
df.columns
```

```python
inputs=df[['address', 'name', 'online_order', 'book_table', 'rate','phone', 'location', 'rest_type', 'dish_liked', 'cuisines',
       'approx_cost(for two people)']]
```

```python
inputs.location.unique()
```

```python
print("Adding filter to user choices")
print("1. Online order Availability ")
print("2. Booking order Availability ")
print("3. Rating out of 5 ")
print("4. Neighbourhood Location")
print("5. Restaurant type ")
print("6. Dishes Liked ")
print("7.Cuisines")
print("8. Approxiamte cost for 2 people")
print("Enter 0 to finish filtering Choices\n\n")

final=pd.DataFrame()
final=inputs
flag=True

while(flag):
    n=int(input())
    if n==0:
        flag=False
    elif n==1:
        final['Online_order']=inputs['online_order']
    elif n==2:
        final['Book_table']=inputs['book_table']
    elif n==3:
        final['Rating']=inputs['rate']
    elif n==4:
        final['Location']=inputs['location']
    elif n==5:
        final['Rest_type']=inputs['rest_type']
    elif n==6:
        final['Dish_liked']=inputs['dish_liked']
    elif n==7:
        final['Cuisines']=inputs['cuisines']
    elif n==8:
        final['Apporxi']=inputs['approx_cost(for two people)']
        
final
```

```python
filtered=final
for column in final:
    if column=="Online_order":
        print("Online order Availability - Yes/No ")
        online_choices=str(input())
        filtered=filtered[filtered['Online_order']==online_choices]
        
    if column=="Book_table":
        print("Booking Table Availability - Yes/No ")
        booking_choices=str(input())
        filtered=filtered[filtered['Book_table']==booking_choices]
        
    if column=="Rating":
        print("Enter min and max rating out of 5")
        mini=str(input("Min_Rate : "))
        maxi=str(input("Max_Rate : "))
        filtered=filtered[(filtered['Rating']>mini) &(filtered['Rating']<maxi)]
        
    if column=="Location":
        print("Enter the location of Restaurant Preferred")
        location_choice=str(input())
        filtered=filtered[filtered['Location']==location_choices]
        
    if column=="Dish_liked":
        print("Choose your dish : ")
        dish_choice=str(input())
        filtered=filtered[filtered[filtered.Dish_liked.apply(lambda x:dish_choice in str(x))]]
        
    if column=="Cuisines":
        print("Choose your Cuisines")
        cui_choices=str(input())
        filtered=filtered[filtered.Cuisines.apply(lambda x:cui_choices in str(x))]
        
    if column=="Apporxi":
        print("Enter Your budget")
        cost=str(input())
        filtered=filtered[filtered['Apporxi']<cost]
        
        
filtered
```
